```python
-- SQL SUBQUERIES
-- subqueries with from clause
-- Q. Find the products that have a price higher than the average price of all products
SELECT
*
FROM(
    SELECT
        ProductID,
        Price,
        AVG(Price) OVER() avgprice
    FROM Sales.Products
)t WHERE Price > avgprice
```

```python
-- Rank the customers based on their total amount of sales 
SELECT
*,
RANK() over(order by sales DESC) AS salesrank
FROM
(
    SELECT
        customerid,
        SUM(sales) totalsales
    from sales.orders
    GROUP BY customerid
)t

```

```python
SELECT
*,
RANK() OVER(ORDER BY totalsales DESC) as salesrank
FROM(
    SELECT
            customerid,
            SUM(sales) totalsales
        from sales.orders
        GROUP BY customerid
) t
```

```python
-- Q. Show the product ids, names, prices and total number of orders
-- scalar subquery
SELECT
    productid,
    product,
    price,(
            SELECT
            count(*) 
            from sales.orders
    )totalorders 
from sales.products
```

```python
-- join subquery
-- Q. show all customer details and find the total orders of each customer

SELECT 
c.*,
o.totalorders 
FROM Sales.Customers AS c 
LEFT JOIN (
    SELECT
    customerid,
    COUNT(*) totalorders
    FROM Sales.Orders
    GROUP BY CustomerID) AS o  
ON c.CustomerID = o.CustomerID
```

```python
-- find the products that have price higher than the average price of all products
SELECT 
    ProductID,
    price
from Sales.products
WHERE Price > (
                SELECT
                    AVG(Price) as aerageprice
                    FROM Sales.Products
)
```

```python
-- show the details of orders made by customers in germany
-- IN operator 
SELECT
*
FROM Sales.Orders
WHERE CustomerID IN (
        SELECT
        CustomerID
        from Sales.Customers
        WHERE Country = 'germany'
)
```

```python
-- Using ANY operator
-- find the female employees whose salaries are greater thatn the salaries of any male employees

SELECT
    EmployeeID,
    FirstName,
    Salary
FROM Sales.Employees 
WHERE Gender = 'F'
AND Salary > ANY(
                    SELECT 
                        Salary
                    FROM Sales.Employees
                    WHERE  Gender = 'M'
)
```
