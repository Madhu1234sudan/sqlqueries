# CASE Statement

```python
SELECT
    first_name,
    salary,
    case 
        when salary >= 80000 then 'High'
        ELSE 'Low'
    END AS salary_distribution
FROM Employees
```

```python
-- Q2 
SELECT
    product_name,
    price,
    case 
        when price >= 50000 then 'Expensive'
        when price >= 20000 then 'Moderate'
        else 'cheap'
    END as price_level
FROM Products
```
