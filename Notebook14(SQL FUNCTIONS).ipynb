# String Functions


## 1. Concat

```python
/* CONCAT — Q1
From Employees:
Display full employee name using:
first_name
last_name */

SELECT 
    CONCAT(first_name, last_name) AS full_name
FROM Employees
```

```python
/* CONCAT — Q2
From Customers:
create:
customer_info
format:
CustomerName - City */
SELECT 
    CONCAT(customer_name, ' -', ' ',  city) AS customer_info 
FROM Customers
```

```python
/* CONCAT — Q3
From Products:
Create:
product_label
Format:
ProductName (Category)*/

SELECT
    CONCAT(product_name, ' (', category, ')') as ProductName
FROM Products
```

## 2. Upper()

```python
/* UPPER — Q1
From Employees:
Display:
first_name in uppercase
Alias:
upper_name */

SELECT 
    UPPER(first_name) as upper_name
FROM Employees
```

## 3. Lower()


```python
SELECT
    LOWER(category) AS lower_category
FROM Products
```

## 4. Trim()

```python
SELECT
    TRIM(customer_name) as clean_name
FROM Customers
```

## Replace() 

```python
SELECT
    REPLACE(category, 'Electronics', 'Electronic Items')
FROM Products
```

# Calculation Functions

## 1. LEN()

```python
SELECT
    LEN(first_name) AS name_length
FROM Employees
```

## 2. LEFT()

```python
SELECT
    LEFT(product_name, 3) AS short_name
FROM Products
```

## RIGHT()

```python
SELECT
    RIGHT(customer_name, 3) AS last_chars 
FROM Customers
```
