## 1. Year()

```python
SELECT
    order_id,
    YEAR(order_date) AS order_year
FROM Orders
```

## 2. MONTH()

```python
SELECT
    order_id,
    MONTH(order_date) AS order_month
FROM Orders
```

## 3. DAY()

```python
SELECT
    order_id,
    DAY(order_date) as order_day
FROM Orders
```

## 4. DATEPART()

```python
SELECT
    order_id,
    DATEPART(QUARTER, order_date) AS order_quarter
FROM Orders
```

## 5. DATENAME()

```python
SELECT
    order_id,
    DATENAME(MONTH, order_date) as month_name
FROM Orders
```

```python
SELECT
    order_id,
    EOMONTH(order_date) AS month_end
FROM Orders
```

```python
SELECT
    order_id,
    order_date,
    DATEADD(DAY, 10, order_date) AS new_date
FROM Orders
```
