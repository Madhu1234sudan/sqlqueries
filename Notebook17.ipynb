# Null Functions

## 1. ISNULL(): It will replace NULL with another value.

```python
SELECT
    first_name,
    manager_id,
    ISNULL(manager_id, 0) manager
FROM Employees
```

## 2. COALESCE(): Returns first non-NULL value.

```python
SELECT
    COALESCE(null, NULL, 'SQL') AS result
```

## 3. NULLIF(): Returns NULL if two values are equal.

```python
SELECT
    NULLIF(100, 100) AS result
```

## 4. IS NULL(): Checks whether value is NULL.

```python
SELECT
    first_name,
    manager_id
FROM Employees
where manager_id is null 
```

## 5. IS NOT NULL: Checks whether value is NOT NULL.


```python
SELECT
    first_name,
    manager_id    
FROM employees
where manager_id is not  null
```
