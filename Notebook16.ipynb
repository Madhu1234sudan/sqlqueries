## 8. CAST()

```python
SELECT
    CAST(GETDATE() AS date ) AS converted_date
```

## 9.CONVERT(): IT CAN CHANGE THE FORMAT OF THE DATE


```python
SELECT
    CONVERT(varchar, GETDATE(), 103) AS formatted_date
```

## 10. GETDATE(): Gives current date and time

```python
SELECT
    GETDATE() as current_datetime
```

## 11. DATETRUNC(): Truncates date to beginning of specified part

```python
SELECT
    order_id,
    DATETRUNC(MONTH, order_date ) AS month_start,
FROM orders
```
