# Subqueries : 
    - 1. single-row subqueries

Single-Row Subquery — Q1

From Employees:

Display:

first_name
salary

Show employees whose salary is greater than average salary of all employees.

Order by salary descending.


```python
SELECT
    first_name,
    salary
FROM Employees
WHERE salary > (select 
                AVG(salary) AS avg_salary
                FROM Employees
                ) ORDER BY salary DESC
```

Single-Row Subquery — Q2

From Employees:

Display:

first_name
salary

Show employee(s) having highest salary.


```python
SELECT
    first_name,
    salary
FROM Employees
WHERE salary = (select 
                MAX(salary)
                FROM Employees)
```


    -2. Multi-row subqueries

Multi-Row Subqueries
These return:
multiple rows

** So operators like:
**=**
**>**
**<**
** usually fail. **
Instead use:
IN
ANY
ALL
EXISTS

Multi-Row Subquery — Q1

From Employees:

Display:

first_name
department_id

Show employees working in departments located in:

Mumbai

```python
SELECT
    first_name,
    department_id
FROM Employees
WHERE department_id IN(select 
                        department_id
                        FROM Departments
                        WHERE [location] = 'Mumbai')
```


    -3. Correlated Subqueries

Normal subquery:

runs once

Correlated subquery:

runs for every outer row

Inner query depends on outer query.

<span style="color: red;">Correlated Subquery — Q1

From Employees:

Display:

first_name
department_id
salary

Show employees whose salary is greater than average salary of their own department.</span>


```python
SELECT
    e.first_name,
    e.department_id,
    e.salary
FROM Employees e 
WHERE salary > (select 
                avg(salary)
                FROM Employees
                WHERE department_id = e.department_id)
```

    -** Exist in subquery
Purpose:
Checks whether related rows exist.

Returns:

TRUE or FALSE

If subquery finds at least one row:

EXISTS = TRUE

<span style="color: red;">EXISTS — Q1

From Customers:

Display:

customer_name

Show customers who have placed at least one order.</span>

```python
SELECT
    c.customer_name
FROM Customers c
WHERE exists (select 1
              FROM Orders o  
              where c.customer_id = o.customer_id )
```

<span style="color:red">EXISTS — Q2

From Products:

Display:

product_name

Show products that exist in:
OrderDetails

Meaning:

products that were ordered at least once</span>

```python
SELECT
    p.product_name
FROM Products p 
WHERE EXISTS (SELECT 1
                FROM OrderDetails od 
                WHERE p.product_id = od.product_id)
```

    -** NOT EXISTS

Opposite of EXISTS.

Very important in interviews.

NOT EXISTS

Purpose:
Checks rows that do NOT have related records.

<span style="color:red">NOT EXISTS — Q1

From Employees:

Display:

first_name

Show employees who are NOT assigned to any project in:

EmployeeProjects</span>

```python
SELECT
    e.first_name
FROM Employees e 
where not exists(SELECT 1
                    from employeeprojects ep 
                        where e.employee_id = ep.employee_id )
```

    -4.Subquery in FROM Clause

Also called:

Derived Table

Very important for:

intermediate calculations
layered aggregations
cleaner complex SQL
Core Idea

Subquery behaves like:

temporary table

<span style="color:red">Subquery in FROM — Q1

From Employees:

Create department-wise average salary using subquery.

Then display:

department_id
avg_salary

from derived table alias:

dept_avg</span>

```python
SELECT * 
FROM(SELECT 
        department_id,
        AVG(salary) avg_salary
        FROM Employees
        GROUP BY department_id 
) AS dept_avg
```
