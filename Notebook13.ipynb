```python
/* SELF JOIN Question 4 — Manager Count

From the Employees table:

Write a query to display:

manager first name
number of employees reporting to that manager

Requirements:

show only managers who have at least one employee
order by employee count descending

Write the query.*/

SELECT 
    m.first_name AS manager_firstname,
    COUNT(e.employee_id) AS report_count
FROM Employees e
INNER JOIN Employees m 
ON e.manager_id = m.employee_id
GROUP BY m.first_name
ORDER BY report_count DESC

```

```python
/* Easy JOIN Practice — Question 1

From the Employees and Departments tables:

Write a query to display:

employee first name
employee salary
department name

Requirements:

use INNER JOIN
show only employees whose salary is greater than 70000
order by salary descending

This is straightforward filtering + join.

Write the query. */

SELECT 
    e.first_name as employee_firstname,
    e.salary,
    d.department_name
FROM Employees AS e
INNER JOIN Departments AS d   
ON e.department_id = d.department_id
WHERE salary > 70000
ORDER BY salary DESC
```

```python
/* Easy JOIN Practice — Question 2

From the Employees and Departments tables:

Write a query to display:

employee first name
department name
department location

Requirements:

show only employees working in the IT department
order by employee first name ascending

Write the query. */

SELECT 
    e.first_name as employee_firstname,
    d.department_name,
    d.location 
FROM Employees e
RIGHT JOIN Departments as d  
ON e.department_id = d.department_id
WHERE d.department_name = 'IT'
ORDER BY e.first_name ASC
```

```python
/* JOIN Practice — Question 3

From the Employees and Departments tables:

Write a query to display:

employee first name
salary
department name

Requirements:

show ALL employees even if department is missing
use the correct JOIN
order by employee first name ascending

Before writing the query, ask yourself:

Which table's records must always appear?

That answer determines the JOIN type.*/

SELECT 
    e.first_name,
    e.salary,
    d.department_name
FROM Employees AS e
LEFT JOIN Departments AS d    
ON e.department_id = d.department_id
ORDER BY e.first_name ASC
```

```python
/* OIN Practice — Question 4

From the Departments and Employees tables:

Write a query to display:

department name
employee first name

Requirements:

show ALL departments even if no employees exist
order by department name ascending

This question checks whether you truly understood LEFT vs RIGHT JOIN.

Write the query.*/

SELECT 
    e.first_name,
    d.department_name
FROM Employees e 
LEFT JOIN Departments d    
ON e.department_id = d.department_id
ORDER BY d.department_name ASC
```

```python
/* JOIN Practice — Question 5

From the Customers and Orders tables:

Write a query to display:

customer name
order_id
order_date
total_amount

Requirements:

show ALL customers even if they have not placed any orders
order results by customer name ascending

Write the query.*/
SELECT 
    c.customer_name,
    o.order_id,
    o.order_date,
    o.total_amount
FROM CUSTOMERS AS c 
LEFT JOIN Orders as o 
ON c.customer_id= o.customer_id
```

```python
/* JOIN Practice — Question 6

From the Orders, OrderDetails, and Products tables:

Write a query to display:

order_id
product_name
category
quantity

Requirements:

show only orders where quantity is greater than 1
use appropriate JOINs
order by order_id ascending

This is your first 3-table JOIN question.

Think step-by-step:

Orders
   ↓
OrderDetails
   ↓
Products

Write the query. */

SELECT 
    o.order_id,
    od.quantity,
    p.product_name,
    p.category 
FROM Orders o  
LEFT JOIN orderdetails od
ON o.order_id = od.order_id
LEFT JOIN products as p  
ON od.product_id = p.product_id
where od.quantity > 1
order by o.order_id ASC 
```

```python
/* JOIN Practice — Question 7

From these tables:

Customers
Orders
OrderDetails
Products

Write a query to display:

customer name
order_id
product name
quantity
total_amount

Requirements:

show only products belonging to the 'Electronics' category
show all customers even if they have no orders
order by customer name ascending

Think carefully:

Customers
   ↓
Orders
   ↓
OrderDetails
   ↓
Products

Main question:

Which table must always stay?

That decides the first JOIN type */

SELECT 
    c.customer_name,
    o.order_id,
    o.total_amount,
    od.quantity,
    p.product_name,
    p.category,
    p.product_id
FROM customers c 
LEFT JOIN orders o
ON c.customer_id = o.customer_id
LEFT JOIN orderdetails as od 
ON o.order_id = od.order_id
LEFT JOIN products p 
ON od.product_id = p.product_id
AND p.category = 'Electronics'
ORDER BY c.customer_name ASC
```
