# Window Functions

## 1. Row Number(): Assign unique sequential number to rows. And it does not take any argument.

```python
SELECT
    first_name,
    salary,
    ROW_NUMBER() OVER(ORDER BY salary DESC) AS row_num 
FROM Employees
```

## 2. Rank(): Assigns rank with the gaps for ties. it does not take any argument.

```python
SELECT
    first_name,
    salary,
    RANK() OVER (ORDER BY salary DESC) AS salary_rank
FROM Employees
```

## 3. DENSE_RANK(): Assigns rank without gaps for ties. and it does not take any argument.

```python
SELECT
    first_name,
    salary,
    DENSE_RANK() OVER (ORDER BY salary DESC) as dense_rank_num
FROM Employees
```

## 4. LEAD():  Gets next row value without self join. and it takes an argument as an expression(column)

```python
SELECT
    first_name,
    salary,
    LEAD(salary) OVER (ORDER BY salary DESC) AS next_salary
FROM Employees
```

## 5. LAG(): Gets previous row value. it takes one argument as an expression(column)

```python
SELECT
    first_name,
    salary,
    LAG(salary) OVER (ORDER BY salary DESC) AS prev_salary 
FROM Employees
```

## 6. NTILE(): Divides rows into equal groups or bukets. It will take strictly numeric(number)

```python
SELECT
    first_name,
    salary,
    NTILE(3) OVER (ORDER BY salary DESC) AS salary_group 
FROM Employees
```

## 7. FIRST_VALUE(): Returns first value in window. It will take exactly one argument as an expression(column)

```python
SELECT
    first_name,
    salary,
    FIRST_VALUE(salary) OVER (ORDER BY salary DESC) AS highest_salary
FROM Employees
```

## 8. LAST_VALUE(): Returns last value in window

```python
SELECT
    first_name,
    salary,
    LAST_VALUE(salary) OVER ( ORDER BY salary desc
    ROWS BETWEEN unbounded preceding and unbounded following) as lowest_salary
FROM Employees
```

## 9. OVER() Clause: Defines windows for window functions

```python
SELECT
    first_name,
    salary,
    SUM(salary) OVER () total_salary
FROM Employees
```

## 10. PARTITION BY(): Divides rows into groups for window functions without collapsing rows

```python
SELECT
    first_name,
    department_id,
    salary,
    DENSE_RANK() OVER (PARTITION BY department_id ORDER BY salary DESC) AS dept_rank
FROM Employees
```
