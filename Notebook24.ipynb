<span style="color:red">1.	Which channel is most frequently used for transactions?</span>

```python
SELECT 
    top 1 
    Store_type,
    COUNT(*) as channel 
FROM Transactions
GROUP BY Store_type
ORDER BY channel DESC
```

```python
with channel AS(
    SELECT 
        Store_type,
        COUNT(*) AS most_frequent_transactions
    FROM Transactions
    GROUP BY Store_type
)
SELECT *
FROM channel
WHERE most_frequent_transactions = (select 
                                        MAX(most_frequent_transactions) 
                                    FROM channel)
```

```python
with channel_frequency AS(

        SELECT 
            Store_type,
            COUNT(*) AS most_frequent_transactions
        FROM Transactions
        GROUP BY Store_type
)
SELECT * 
FROM channel_frequency
WHERE most_frequent_transactions = (select 
                                    MAX(most_frequent_transactions) 
                                    FROM channel_frequency)
```

```python
WITH channel_frequency AS(
    SELECT
        Store_type,
        COUNT(*) AS least_frequent_transactions
    FROM Transactions
    GROUP BY Store_type
)
SELECT * 
FROM channel_frequency 
WHERE least_frequent_transactions = (select 
                                        MIN(least_frequent_transactions)
                                    FROM channel_frequency)
```

<span style="color: red">2.	What is the count of Male and Female customers in the database?</span>

```python
SELECT
    Gender,
    COUNT(*) AS gender_count
FROM Customer
WHERE Gender IS NOT NULL
GROUP BY Gender
```

<span style="color:red">4.	How many sub-categories are there under the Books category?</span>

```python
SELECT  
    COUNT(Distinct prod_subcat ) AS sub_categories
FROM prod_cat_info
WHERE prod_cat = 'Books'
```
