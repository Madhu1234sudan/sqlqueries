```python
/* create report showing total sales for each of the following categories:
  High (sales over 50), Medium(sales over 20-50), and low (sales 20 or less)
  sort the categories from highest sales to lowest */

  SELECT 
        category,
        sum(Sales) as totalsales
        from(select 
                orderid,
                sales,
                case 
                    when sales > 50 THEN 'High'
                    when Sales > 20 then 'Medium'
                    else 
                        'Low'
        end category
        from sales.Orders
    )t 
    GROUP BY category
    
    order by totalsales DESC
  
```

```python
select 
    id,
    first_name,
    CASE 
        when score > 700 THEN  'High'
        when score > 400 THEN 'Medium'
        ELSE
            'low'
    END
from customers
```

```python
SELECT
    id,
    first_name,
    sum(score) as totalscore
    from ( select 
                id,
                first_name
                case 
                    when score > 700 then 'high'
                    when score > 500 then 'Medium'
                end
                from customers
        )
    group by first_name
    order by totalscore
```
